```python
import pandas as pd

df = pd.read_excel("C:/Users/MANSI/ad_detector/data/alzheimer_audio_info.xlsx")

# keep only useful columns
df = df[["subject id", "type", "age", "gender"]]

# remove missing rows
df = df.dropna()

# convert labels
df["label"] = df["type"].map({
    "AD": 1,
    "HC": 0
})

print(df.head())
```

```python
audio_root = r"C:\Users\MANSI\ad_detector\data\OneDrive_1_3-17-2026"
```

```python
print(df.columns)
```

```python
texts = []
labels = []

count = 0

for _, row in df.iterrows():
    subject_id = str(row["subject id"]).strip()
    label = row["label"]

    subject_folder = None
    for folder in os.listdir(audio_root):
        if subject_id in folder:
            subject_folder = os.path.join(audio_root, folder)
            break

    if subject_folder:
        for file in os.listdir(subject_folder):
            if file.lower().endswith(".wav"):
                file_path = os.path.join(subject_folder, file)

                try:
                    text = audio_to_text(file_path)

                    texts.append(text)
                    labels.append(label)

                    count += 1
                    print(f"Processed {count}")

                    # Save progress
                    if count % 50 == 0:
                        pd.DataFrame({
                            "text": texts,
                            "label": labels
                        }).to_csv("temp_text.csv", index=False)

                    if count == 400:   # change if needed
                        break

                except Exception as e:
                    print("Error:", e)

        if count == 400:
            break

print("Total processed:", count)
```

```python
# ============================================
# CELL 1 — Imports
# ============================================
import os
import re
import pickle
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_validate
from sklearn.metrics import classification_report, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
print("Imports done ✅")
```

```python
# ============================================
# CELL 2 — Load & Clean Data
# ============================================
df = pd.read_csv(r"C:\Users\MANSI\ad_detector\notebooks\temp_text.csv")
df = df.dropna(subset=["text"])
df = df[df["text"].str.strip() != ""]

def clean_text(text):
    if not isinstance(text, str):
        return ""
    text = text.lower()
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df["text_clean"] = df["text"].apply(clean_text)
df = df[df["text_clean"].str.len() > 0]

print("Total samples:", len(df))
print("AD:", df["label"].sum(), "| HC:", (df["label"]==0).sum())
df.head()
```

```python
# ============================================
# CELL 3 — Train/Test Split
# ============================================
texts  = df["text_clean"].tolist()
labels = df["label"].tolist()

X_train, X_test, y_train, y_test = train_test_split(
    texts, labels,
    test_size=0.2,
    random_state=42,
    stratify=labels
)

print("Train:", len(X_train), "| Test:", len(X_test))
```

```python
# ============================================
# CELL 4 — Train TF-IDF + Logistic Regression
# ============================================
pipeline = Pipeline([
    ("tfidf", TfidfVectorizer(
        max_features=500,
        ngram_range=(1, 2),
        stop_words=None       # Chinese text ke liye None rakhna
    )),
    ("clf", LogisticRegression(
        max_iter=1000,
        class_weight="balanced",
        random_state=42
    ))
])

pipeline.fit(X_train, y_train)
y_pred = pipeline.predict(X_test)

print("=== TF-IDF + Logistic Regression ===")
print(classification_report(y_test, y_pred, target_names=["HC", "AD"]))
```

```python
# ============================================
# CELL 5 — Confusion Matrix
# ============================================
fig, ax = plt.subplots(figsize=(5, 4))
ConfusionMatrixDisplay.from_predictions(
    y_test, y_pred,
    display_labels=["HC", "AD"],
    cmap="Greens", ax=ax
)
ax.set_title("Linguistic Model — Confusion Matrix")
plt.tight_layout()
plt.savefig("confusion_matrix_linguistic.png", dpi=120)
plt.show()
print("Plot saved ✅")
```

```python
# ============================================
# CELL 6 — Cross Validation
# ============================================
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_results = cross_validate(
    pipeline, texts, labels,
    cv=skf,
    scoring=["recall", "f1", "accuracy"]
)

print("=== 5-Fold Cross Validation ===")
for metric in ["recall", "f1", "accuracy"]:
    scores = cv_results[f"test_{metric}"]
    print(f"{metric.capitalize():10s}: {scores.mean():.2f} ± {scores.std():.2f}")
```
