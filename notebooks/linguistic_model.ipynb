```python
import pandas as pd

df = pd.read_excel("C:/Users/MANSI/ad_detector/data/alzheimer_audio_info.xlsx")

# keep only useful columns
df = df[["subject id", "type", "age", "gender"]]

# remove missing rows
df = df.dropna()

# convert labels
df["label"] = df["type"].map({
    "AD": 1,
    "HC": 0
})

print(df.head())
```

```python
audio_root = r"C:\Users\MANSI\ad_detector\data\OneDrive_1_3-17-2026"
```

```python
print(df.columns)
```

```python
texts = []
labels = []

count = 0

for _, row in df.iterrows():
    subject_id = str(row["subject id"]).strip()
    label = row["label"]

    subject_folder = None
    for folder in os.listdir(audio_root):
        if subject_id in folder:
            subject_folder = os.path.join(audio_root, folder)
            break

    if subject_folder:
        for file in os.listdir(subject_folder):
            if file.lower().endswith(".wav"):
                file_path = os.path.join(subject_folder, file)

                try:
                    text = audio_to_text(file_path)

                    texts.append(text)
                    labels.append(label)

                    count += 1
                    print(f"Processed {count}")

                    # Save progress
                    if count % 50 == 0:
                        pd.DataFrame({
                            "text": texts,
                            "label": labels
                        }).to_csv("temp_text.csv", index=False)

                    if count == 400:   # change if needed
                        break

                except Exception as e:
                    print("Error:", e)

        if count == 400:
            break

print("Total processed:", count)
```

```python
df_text = pd.read_csv("temp_text.csv")
df_text = df_text.dropna(subset=["text"])   
texts = df_text["text"].tolist()
labels = df_text["label"].tolist()

print(len(texts))
```

```python
def clean_text(text):
    if isinstance(text, str):   # ✅ check if text is string
        text = text.lower()
        text = re.sub(r'[^a-zA-Z\s]', '', text)
        text = re.sub(r'\s+', ' ', text).strip()
        return text
    else:
        return ""   # replace NaN with empty string

texts = [clean_text(t) for t in texts]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import pandas as pd
import pickle

# Load data
df = pd.read_csv("temp_text.csv")
df = df.dropna(subset=["text"])
texts = df["text"].tolist()
labels = df["label"].tolist()

# Split
X_train, X_test, y_train, y_test = train_test_split(
    texts, labels,
    test_size=0.2,
    random_state=42,
    stratify=labels
)

# TF-IDF pipeline
pipeline = Pipeline([
    ('tfidf', TfidfVectorizer(
        max_features=500,
        ngram_range=(1, 2),
        analyzer='char_wb'  # character level — better for Chinese
    )),
    ('clf', LogisticRegression(
        max_iter=3000,
        class_weight='balanced',
        random_state=42
    ))
])

pipeline.fit(X_train, y_train)
y_pred = pipeline.predict(X_test)

print(classification_report(y_test, y_pred))

import os
os.makedirs("models", exist_ok=True)

with open("models/linguistic_model.pkl", "wb") as f:
    pickle.dump(pipeline, f)
print("Saved!")
# Save
with open("models/linguistic_model.pkl", "wb") as f:
    pickle.dump(pipeline, f)
print("Saved!")
```
